```python
#imports
import pandas as pd
import warnings
from datetime import time
```

```python
warnings.filterwarnings('ignore')
```

```python
#Import the dataset and load in to pandas dataframe
operationsDF = pd.read_csv('operations_2022.csv')
operationsDF
```

```python
operationsDF.info()
```

```python
#convert date column
operationsDF['Date'] = pd.to_datetime(operationsDF['Date'], errors ='coerce')
```

```python
#rename the date and business cost columns
operationsDF.rename(columns={'Date': 'OperationDate', 'BusinessCost':'TotalCost'}, inplace=True)
operationsDF
```

```python
#Export as csv file
operationsDF.to_csv('clean_operations_2022.csv', index=False)
```
